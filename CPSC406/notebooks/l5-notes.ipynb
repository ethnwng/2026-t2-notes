```julia
using LinearAlgebra

A1 = [1 2; 2 4]
A2 = [1 0 0; 0 1 0; 0 0 0]

@show rank(A1)
@show rank(A2)
@show svdvals(A1)
@show svdvals(A2)
```

```julia
A = [3 0 0; 1 2 0]

U, Sig, V =svd(A)

@show U 
@show Sig
@show V 

# best rank 1-approx
A1 = U[:, 1] * Sig[1] * V[:, 1]'
```

```julia
@show norm(A-A1)
@show Sig[2]
```

```julia
u = [1 ; 2; 3]
v = [1; 1]
A = u*v'
@show A
@show rank(A)
@show svdvals(A)
```
