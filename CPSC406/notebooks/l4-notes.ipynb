```julia
A = [3 1; 0 2; 0 0]
```

```julia
U, Sig, V = svd(A)
```

```julia
A_reshape = U*Diagonal(Sig)*V' # Diagonal(Sig) takes the 2x1 vector and makes it into a 2x2 diagonal matrix
```

```julia
@show A'A 
@show sqrt.(eigvals(A'A)) # compare the square root of the eigenvalues of A'A to the singular values
```

```julia
A = [3 1; 0 2]

# decompose A
U, Sig, V = svd(A)
@show Sig
@show A*V[:,1]
@show A*V[:, 2]

# generate circle
theta = range(0, 2pi; length=200)
circle_p = [cos.(theta)'; sin.(theta)']

# apply transformation
ellipse_p = A*circle_p

plot(circle_p[1, :], circle_p[2, :], label="Unit Circle", aspect_ratio=1)
plot!(ellipse_p[1, :], ellipse_p[2, :], label="Transformed Ellipse", aspect_ratio=1)
```
